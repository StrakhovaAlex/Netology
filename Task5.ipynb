```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
  

# p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит; 
def people():
    person_name = input('Введите номер документа ')
    doc_dict = {d['number']:d['name'] for d in documents}
    if person_name in doc_dict.keys():
        print(doc_dict.get(person_name))
    else:
        print('Ошибка ввода данных')
        
        
# s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится        
def shelf():
    document_number = input('Введите номер документа ')
    doc_dict = {d['number']:d['name'] for d in documents}
    if document_number in doc_dict.keys():
        for key , value in directories.items():  
            if document_number in value:
                print(key)
    else:
        print('Ошибка ввода номера документа')

        
#l – list – команда, которая выведет список всех документов в формате
#passport "2207 876234" "Василий Гупкин"

def list():
    for x in documents:
        print('**', x['type'] , ' "', x['number'], '" "', x['name'], '"**')

                
#as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
def add_shelf():
    cell = input('Введите номер новой полки ')
    if cell in directories.keys():
        print('Номер полки существует')
    else:
        directories[cell] = [] 
    print(directories)
    
#d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. 
#Предусмотрите сценарий, когда пользователь вводит несуществующий документ;
def delete(documents,directories):
    doc_number = input('Введите номер документа')
    documents = [doc for doc in documents if not doc['number']==doc_number]
    for shelf,docs in directories.items():
        directories[shelf] = [doc for doc in docs if not doc==doc_number]
    return documents,directories

#m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. 
#Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ
#или переместить документ на несуществующую полку; 
def move(directories):
    doc_number = input('Введите номер документа')
    target_shelf = input('Введите номер полки')
    existing_doc = 0
    for shelf,docs in directories.items():
        if doc_number in docs:
            existing_doc = 1
            if shelf==target_shelf:
                return directories
            else:
                directories[shelf] = [doc for doc in docs if not doc==doc_number]
    if existing_doc:
        if target_shelf in directories.keys():
            directories[target_shelf].append(doc_number)
        else:
            directories[target_shelf] = [doc_number]
    return directories



def add():
    doc_number = input('Введите номер документа')
    doc_type = input('Введите тип документа')
    owner_name = input('Введите имя владельца документа')
    shelf_num = input('Введите номер полки')
    documents.append({'type': doc_type, 'number': doc_number, 'name': owner_name})
    if shelf_num in directories.keys():
        directories[shelf_num].append(doc_number)
    else:
        directories[shelf_num] = doc_number
    print(directories)
    print(documents)

    
 #вызов команды    
def main(documents,directories):
    
    function_number = input('Введите название команды ')
    if function_number == 'p':
        people()
    elif function_number == 's':
        shelf()
    elif function_number == 'l':
        list()
    elif function_number == 'as':
        add_shelf()
    elif function_number == 'd':        
        documents,directories = delete(documents,directories)
        print(documents)
        print(directories)
    elif function_number == 'm':
        directories = move(directories)
        print(directories)
    elif function_number == 'a':
        add()
    else:
        return
main(documents,directories)    
```
